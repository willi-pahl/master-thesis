```python
# Plot activation function.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
inputs = np.arange(-5, 5, 0.1)
```

```python
def identity(x: float) -> float:
    """
    Identity activation function.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    return x


def identity_derivative(x: float) -> float:
    """
    Identity derivative.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    return 1


def binary_step(x: float) -> float:
    """
    Binary step activation function.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    if x < 0:
        return 0
    else:
        return 1


def binary_step_derivative(x: float) -> float:
    """Binary step derivative.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    return 0


def gaussian(x: float) -> float:
    """
    Gaussian activation function.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    return math.exp(-(x * x))


def gaussian_derivative(x: float) -> float:
    """
    Gaussian derivative.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    return -2 * x * math.exp(-(x * x))


def sigmoid(x: float) -> float:
    """
    Sigmoid activation function.

    Args:
        x (float): Input value

    Returns:
        float: Output.
    """
    return 1 / (1 + math.exp(-x))


def sigmoid_derivative(x: float) -> float:
    """
    Sigmoid derivative.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    return sigmoid(x) * (1 - sigmoid(x))


def relu(x: float) -> float:
    """
    Rectified Linear Unit activation function.

    Args:
        x (float): Input value.

    Returns:
        float: Output
    """
    if x < 0:
        return 0
    else:
        return x


def relu_derivative(x: float) -> float:
    """
    Rectified Linear Unit derivative.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    if x < 0:
        return 0
    else:
        return 1


def leaky_relu(x: float) -> float:
    """
    Leaky Rectified Linear Unit activation function.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    alpha: float = 0.2
    if x < 0:
        return x * alpha
    else:
        return x


def leaky_relu_derivative(x: float) -> float:
    """
    Leaky Rectified Linear Unit derivative.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    alpha: float = 0.2
    if x < 0:
        return alpha
    else:
        return 1


def tanh(x: float) -> float:
    """
    Tangens hyperbolic activation function.

    Args:
        x (float): Input value.

    Returns:
        float: Output.
    """
    return math.tanh(x)


def tanh_derivative(x: float) -> float:
    """
    Tangens hyperbolic derivative.

    Args:
        x (float): _description_

    Returns:
        float: _description_
    """
    return 1 - math.tanh(x) ** 2
```

```python
out_identity = []
out_derivation_identity = []
for item in inputs:
    out_identity.append(identity(item))
    out_derivation_identity.append(identity_derivative(item))

plt.figure()
plt.plot(inputs, out_identity, label="Identity")
plt.plot(inputs, out_derivation_identity, color="r", label="Identity derivation")

plt.title("Identity function and its derivation", fontsize=18, fontweight='bold')
plt.xlabel("x", weight="bold", size=16)
plt.ylabel("y", weight="bold", size=16)
plt.grid(True)
plt.legend()
plt.show()
```

```python
out_binary_step = []
out_derivation_binary_step = []
for item in inputs:
    out_binary_step.append(binary_step(item))
    out_derivation_binary_step.append(binary_step_derivative(item))

plt.figure()
plt.plot(inputs, out_binary_step, label="Binary Step")
plt.plot(inputs, out_derivation_binary_step, color="r", label="Binary Step derivation")

plt.title("Binary Step func and its derivation", fontsize=18, fontweight='bold')
plt.xlabel("x", weight="bold", size=16)
plt.ylabel("y", weight="bold", size=16)
plt.grid(True)
plt.legend()
plt.show()
```

```python
out_sigmoid = []
out_derivation_sigmoid = []
for item in inputs:
    out_sigmoid.append(sigmoid(item))
    out_derivation_sigmoid.append(sigmoid_derivative(item))

plt.figure()
plt.plot(inputs, out_sigmoid, label="Sigmoid")
plt.plot(inputs, out_derivation_sigmoid, color="r", label="Sigmoid derivation")

plt.title("Sigmoid function and its derivation", fontsize=18, fontweight='bold')
plt.xlabel("x", weight="bold", size=16)
plt.ylabel("y", weight="bold", size=16)
plt.grid(True)
plt.legend()
plt.show()
```

```python
out_relu = []
out_derivation_relu = []
for item in inputs:
    out_relu.append(relu(item))
    out_derivation_relu.append(relu_derivative(item))


plt.figure()
plt.plot(inputs, out_relu, label="ReLU")
plt.plot(inputs, out_derivation_relu, color="r", label="ReLU derivation")

plt.title("ReLU function and its derivation", fontsize=18, fontweight='bold')
plt.xlabel("x", weight="bold", size=16)
plt.ylabel("y", weight="bold", size=16)
plt.grid(True)
plt.legend()
plt.show()
```

```python
out_leaky_relu = []
out_derivation_leaky_relu = []
for item in inputs:
    out_leaky_relu.append(leaky_relu(item))
    out_derivation_leaky_relu.append(leaky_relu_derivative(item))

plt.figure()
plt.plot(inputs, out_leaky_relu, label="Leaky ReLU")
plt.plot(inputs, out_derivation_leaky_relu, color="r", label="Leaky ReLU derivation")

plt.title("Leaky ReLU func and its derivation", fontsize=18, fontweight='bold')
plt.xlabel("x", weight="bold", size=16)
plt.ylabel("y", weight="bold", size=16)
plt.grid(True)
plt.legend()
plt.show()
```

```python
out_tanh = []
out_derivation_tanh = []
for item in inputs:
    out_tanh.append(tanh(item))
    out_derivation_tanh.append(tanh_derivative(item))

plt.figure()
plt.plot(inputs, out_tanh, label="tanh")
plt.plot(inputs, out_derivation_tanh, color="r", label="tanh derivation")

plt.title("Tangens H func and its derivation", fontsize=18, fontweight='bold')
plt.xlabel("x", weight="bold", size=16)
plt.ylabel("y", weight="bold", size=16)
plt.grid(True)
plt.legend()
plt.show()
```

```python
out_gaussian = []
out_derivation_gaussian = []
for item in inputs:
    out_gaussian.append(gaussian(item))
    out_derivation_gaussian.append(gaussian_derivative(item))

plt.figure()
plt.plot(inputs, out_gaussian, label="gaussian")
plt.plot(inputs, out_derivation_gaussian, color="r", label="gaussian derivation")

plt.title("Gaussian func and its derivation", fontsize=18, fontweight='bold')
plt.xlabel("x", weight="bold", size=16)
plt.ylabel("y", weight="bold", size=16)
plt.grid(True)
plt.legend()
plt.show()
```
